```python
from pycatan.game import Game
from pycatan.statuses import Statuses
from pycatan.card import ResCard, DevCard
import math

import random
```

```python
def rand_city(game, r_points,player):    
    random.shuffle(r_points)
    for i in [i.point.position for i in game.board.get_buildings() if i.type == 0]:
        cit = game.add_city(*i, player.num)
        if cit == 3:                    
            for card in [[ResCard.WHEAT], [ResCard.ORE]]:
                if card == [ResCard.WHEAT]:
                    minn = 3
                if card == [ResCard.ORE]:
                    minn = 2
                if not player.cards.count(card[0]) >= minn:
                    res = [ResCard.WOOD, ResCard.BRICK, ResCard.WHEAT, ResCard.ORE, ResCard.SHEEP]
                    res.remove(card[0])
                    random.shuffle(res)
                    boo = False
                    done = False
                    for card2 in res:
                        if card2 == ResCard.WHEAT:
                            while player.cards.count(card2) >= 7 and not done:
                                (game.trade_to_bank(0, [card2]*4, card[0]))
                                cit = game.add_city(*i, player.num)
                                if cit == 2:
                                    done=True

                        elif card2 == ResCard.ORE:
                            while player.cards.count(card2) >= 6 and not done:
                                (game.trade_to_bank(0, [card2]*4, card[0]))
                                cit = game.add_city(*i, player.num)
                                if cit == 2:
                                    done=True

                        else:
                            while player.cards.count(card2) >= 4 and not done:
                                (game.trade_to_bank(0, [card2]*4, card[0]))
                                cit = game.add_city(*i, player.num)
                                if cit == 2:
                                    done=True
                    if done:
                        break
    return game
```

```python
def rand_settle(game, r_points,player):
    random.shuffle(r_points)                   
    for i in r_points:
        sett = game.add_settlement(player.num,*i)
        if sett == 3:
            for card in [[ResCard.WOOD], [ResCard.BRICK], [ResCard.WHEAT], [ResCard.SHEEP]]:
                if not player.has_cards(card):
                    res = [ResCard.WOOD, ResCard.BRICK, ResCard.WHEAT, ResCard.ORE, ResCard.SHEEP]
                    res.remove(card[0])
                    random.shuffle(res)
                    boo = False
                    done = False
                    for card2 in res:
                        if card2 in [ResCard.WOOD, ResCard.BRICK, ResCard.WHEAT, ResCard.SHEEP]:
                            while player.cards.count(card2) >= 5 and not done:
                                (game.trade_to_bank(0, [card2]*4, card[0]))
                                sett = game.add_settlement(player.num,*i)
                                if sett == 2:
                                    done=True
                                    
                        else:
                            while player.cards.count(card2) >= 4 and not done:
                                (game.trade_to_bank(0, [card2]*4, card[0]))
                                sett = game.add_settlement(player.num,*i)
                                if sett == 2:
                                    done=True
                                    
                    if done:
                        break
    return game
```

```python
def rand_road(game, r_points, player):  
    random.shuffle(r_points)       
    pts = []
    for i in r_points:
        for j in game.board.points[i[0]][i[1]].connected_points:
            j = j.position
            pts.append(tuple(sorted((i, tuple(j)))))
    for i, j in list(set(pts)):
        road = (game.add_road(player.num, list(i), list(j)))
        if road == 3:
            for card in ([[ResCard.WOOD], [ResCard.BRICK]]):
                if not player.has_cards(card):
                    res = [ResCard.WOOD, ResCard.BRICK, ResCard.WHEAT, ResCard.ORE, ResCard.SHEEP]
                    res.remove(card[0])
                    random.shuffle(res)
                    done=False
                    for card2 in res:
                        if card2 in [ResCard.WOOD, ResCard.BRICK]:
                            while player.cards.count(card2) >= 5 and not done:
                                (game.trade_to_bank(0, [card2]*4, card[0]))
                                road = (game.add_road(player.num, list(i), list(j)))
                                if road == 2:
                                    done=True

                        else:
                            while player.cards.count(card2) >= 4 and not done:
                                (game.trade_to_bank(0, [card2]*4, card[0]))
                                road = (game.add_road(player.num, list(i), list(j)))
                                if road == 2:
                                    done=True


                    if done:
                        break
    return game
```

```python
def rand_dev(game, player):
    dev = game.build_dev(player.num)
    if dev == 3:
        for card in ([[ResCard.WHEAT], [ResCard.ORE], [ResCard.SHEEP]]):
                if not player.has_cards(card):
                    res = [ResCard.WOOD, ResCard.BRICK, ResCard.WHEAT, ResCard.ORE, ResCard.SHEEP]
                    res.remove(card[0])
                    random.shuffle(res)
                    done=False
                    for card2 in res:
                        if card2 in [ResCard.WHEAT, ResCard.ORE, ResCard.SHEEP]:
                            while player.cards.count(card2) >= 5 and not done:
                                (game.trade_to_bank(0, [card2]*4, card[0]))
                                dev = game.build_dev(player.num)
                                if dev == 2:
                                    done=True

                        else:
                            while player.cards.count(card2) >= 4 and not done:
                                (game.trade_to_bank(0, [card2]*4, card[0]))
                                dev = game.build_dev(player.num)
                                if dev == 2:
                                    done=True


                    if done:
                        break
    return game
```

```python
def use_dev(game, cards, player,r_points):
    for card in cards:
        print(card)
        if card == DevCard.ROAD:
        # checks the correct arguments are given
        
            road_points = list(set(r_points + [tuple(j.position) for i in game.board.points for j in i if j.building is not None and j.building.type in [0, 2] and j.building.owner == player.num]))
            random.shuffle(road_points)
            for orig_pt in road_points:
                
                dest_1_points = game.board.points[orig_pt[0]][orig_pt[1]].connected_points
                for dest_1 in dest_1_points:
                    dest_1 = dest_1.position
                    orig_2 = [i.position for i in game.board.points[dest_1[0]][dest_1[1]].connected_points]
                    orig_2.remove(list(orig_pt))
                    for orig_2_pt in orig_2:
                    
                        dest_2 = [i.position for i in game.board.points[orig_2_pt[0]][orig_2_pt[1]].connected_points]
                        dest_2.remove(dest_1)
                        for dest_2_pt in dest_2:
                            args = {
                                'road_one': {
                                    'start': orig_pt,
                                    'end': dest_1
                                },
                                'road_two': {
                                    'start': orig_2_pt,
                                    'end':dest_2_pt
                                }
                            }
                            res = game.use_dev_card(player.num, card, args)
                            if res == 2:
                                return game
   
        elif card == DevCard.KNIGHT:
            # checks there are the right arguments



            points = [j.position for i in game.board.points for j in i if j.building is not None and j.building.type in [0, 2]]
            try:
                points.remove(game.board.robber)
            except:
                pass
            rob_point = random.choice(points)
            victim = game.board.points[rob_point[0]][rob_point[1]].building.owner
            args = {
                'robber_pos': rob_point,
                'victim':victim
            }
            print(game.use_dev_card(player.num, card, args))


        elif card == DevCard.MONOPOLY:
            # gets the type of card
            args = {'card_type': random.choice([ResCard.WHEAT, ResCard.ORE, ResCard.WOOD, ResCard.BRICK, ResCard.SHEEP])}
            print(game.use_dev_card(player.num, card, args))

        elif card == DevCard.VP:
            # players do not play developement cards, so it returns an error
            pass

        elif card == DevCard.YOP:
            # checks the player gave two development cards
            args = {'card_one':random.choice([ResCard.WHEAT, ResCard.ORE, ResCard.WOOD, ResCard.BRICK, ResCard.SHEEP]),
                    'card_two':random.choice([ResCard.WHEAT, ResCard.ORE, ResCard.WOOD, ResCard.BRICK, ResCard.SHEEP])}
            print(game.use_dev_card(player.num, card, args))
    return game
```

```python
game = Game(num_of_players=1)
while True:
    try:
        for i in range(2, 13):
            game.add_yield_for_roll(i)
        break
    except:
        continue
last=[None] * len(game.players)
pointss = []


for i in range(len(game.board.points)):
    for j in range(len(game.board.points[i])):
        pointss.append((i,j))

for p in game.players + list(reversed(game.players)):

    while True:
        row, index = random.choice(pointss)

        res = game.add_settlement(player=p.num, r=row, i=index, is_starting=True)
        if res != Statuses.ALL_GOOD:

            continue
        points = game.board.points[row][index].connected_points
        while True:
            res = game.add_road(player=p.num, start=[row, index], end=random.choice(points).position, is_starting=True)
            if res == Statuses.ALL_GOOD:
                break
        last[p.num]= True
        break
if last[p.num]:
    for hexx in game.board.get_hex_indexes_for_point(row, index):
        game.players[p.num].cards.append(game.board.get_card_from_hex(game.board.hexes[hexx[0]][hexx[1]].type))
while max([i.get_VP() for i in game.players]) < 10:   
    for player in game.players:
        roll = game.get_roll()
        if roll == 7:
            points = [j.position for i in game.board.points for j in i if j.building is not None and j.building.type in [0, 2]]
            try:
                points.remove(game.board.robber)
            except:
                pass
            rob_point = random.choice(points)
            victim = game.board.points[rob_point[0]][rob_point[1]].building.owner
            rob = game.move_robber(*rob_point, player.num, victim)
            for player in game.players:
                if len(game.players[player.num].cards) > 7:

                    random.shuffle(game.players[player.num].cards)
                    game.players[player.num].cards = game.players[player.num].cards[:len(game.players[player.num].cards)//2]
        else:
            game.add_yield_for_roll(roll)
        r_points=[]
        for road in game.board.roads:
            r_points.extend([road.point_one, road.point_two])
        r_points = list(set([tuple(i) for i in r_points]))
        funcs = [rand_city, rand_settle, rand_road, rand_dev, use_dev]
        random.shuffle(funcs)
        for func in funcs:
            if func == rand_dev:
                game = func(game, player)
            elif func == use_dev:
                if game.players[player.num].dev_cards:
                    cards = random.choice(list(range(1,len(game.players[player.num].dev_cards)+1)))
                    card_copy = [i for i in game.players[0].dev_cards if i != DevCard.VP]
                    cards = card_copy[:cards]
                    game = func(game, cards, player, r_points)

            else:
                game = func(game, r_points, player)  
print('done')
```
